```python
import pandas as pd
import prplot as prc
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/refs/heads/master/titanic.csv')
style = prp.style.Grayscale()

(prc.chart()
    .data(data)
    .bind(
        x='Embarked',
    )
    .style(style)
    .label(
        title='Embarked Count',
        x='Embarked?',
        y='People Count'
    )
    .plot(prp.plot.Bar())
    .save('bar.png')
    .show()
)
```
